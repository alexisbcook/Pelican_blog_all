```python
# setup
from scipy.stats import beta
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline

def prior_and_posterior_plot(a,b):
    x_prior = np.linspace(beta.ppf(0.01,a,b),
                    beta.ppf(0.99,a,b),100)
    y_prior = beta.pdf(x_prior,a,b)
    max_x_prior = x_prior[y_prior.argmax()]
    x_post = np.linspace(beta.ppf(0.01,a+995,b+5),
                    beta.ppf(0.99,a+995,b+5),100)
    y_post = beta.pdf(x_post,a+995,b+5)
    max_x_post = x_post[y_post.argmax()]

fig, ax = plt.subplots(3, 2, sharex = True, sharey = False, figsize = (18,18))

# alpha = beta = 1
a = 1
b = 1
x_prior, y_prior, max_x_prior, x_post, y_post, max_x_post = prior_and_posterior_plot(a,b)
ax[0,0].plot(x_prior,y_prior)
ax[0,0].set_title(r"Prior $P(\theta)$ for $\alpha$ = %r, $\beta$ = %r; $\theta \sim Unif(0,1)$" %(a,b), fontsize=15)
ax[0,1].plot(x_post,y_post)
ax[0,1].set_title(r"Posterior $P(\theta|D)$ for $\alpha$ = %r, $\beta$ = %r; $\theta_{MAP}$ = %r" %(a,b,max_x_post), fontsize=15)

# alpha = beta = 100
a = 100
b = 100
x_prior, y_prior, max_x_prior, x_post, y_post, max_x_post = prior_and_posterior_plot(a,b)
ax[1,0].plot(x_prior,y_prior)
ax[1,0].set_title(r"Prior $P(\theta)$ for $\alpha$ = %r, $\beta$ = %r; $\hat{\theta}$ = %r" %(a,b,max_x_prior), fontsize=15)
ax[1,1].plot(x_post,y_post)
ax[1,1].set_title(r"Posterior $P(\theta|D)$ for $\alpha$ = %r, $\beta$ = %r; $\theta_{MAP}$ = %r" %(a,b,max_x_post), fontsize=15)

# alpha = 1, beta = 990
a = 1
b = 990
x_prior, y_prior, max_x_prior, x_post, y_post, max_x_post = prior_and_posterior_plot(a,b)
ax[2,0].plot(x_prior,y_prior)
ax[2,0].set_title(r"Prior $P(\theta)$ for $\alpha$ = %r, $\beta$ = %r; $\hat{\theta}$ = %r" %(a,b,max_x_prior), fontsize=15)
ax[2,1].plot(x_post,y_post)
ax[2,1].set_title(r"Posterior $P(\theta|D)$ for $\alpha$ = %r, $\beta$ = %r; $\theta_{MAP}$ = %r" %(a,b,max_x_post), fontsize=15)

plt.show()
```
